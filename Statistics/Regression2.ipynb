# 다중선형회귀분석

```python
import os
import pandas as pd
import numpy as np
import statsmodels.api as sm
```

```python
boston = pd.read_csv('./Boston_house.csv')
boston.head()
```

```python
boston_data = boston.drop(['Target'],axis=1)
boston_data.head()
```

```python
target = boston[['Target']]
x_data = boston[['CRIM','RM','LSTAT']]
x_data.head()
```

```python
x_data1 = sm.add_constant(x_data,has_constant='add')
```

```python
multi_model1 = sm.OLS(target,x_data)
fitted_multi_model1 = multi_model1.fit()
```

```python
fitted_multi_model1.summary()
```

### VIR를 통한 다중공선성확인

```python
from statsmodels.stats.outliers_influence import variance_inflation_factor
```

```python
vif = pd.DataFrame()  # 빈데이터프레임 만들어준다
vif['VIF Factor']=[variance_inflation_factor(x_data2.values,i) for i in range(x_data2.shape[1])]
vif['features']=x_data2.columns
vif

# NOX변수가 VIF가 가장크다   10이상이면 다중공선성이 있을것 같다라고 예상
# 굉장히 많은 다중공선성이 있는것같다라고 보지만  NOX만 제거해보고 분석해보자
```

```python
vif = pd.DataFrame()
x_data2_1 = x_data2.drop(['NOX'],axis=1)
vif['VIF Factor']=[variance_inflation_factor(x_data2_1.values,i) for i in range(x_data2_1.shape[1])]
vif['features']=x_data2_1.columns
vif
```

VIF만 본다면 RM도 10이상의 값을 보여서 제거해야되는 것이 맞지만
위에서 상관행렬을 보았을 떄 큰 문제를 보이지 않으므로 이런변수는 분석자의 인사이트가
필요하다(넣을지 말지)

```python
# 그렇지만 RM을 한번 제거해보자
vif = pd.DataFrame()
x_data2_2 = x_data2_1.drop(['RM'],axis=1)
vif['VIF Factor']=[variance_inflation_factor(x_data2_2.values,i) for i in range(x_data2_2.shape[1])]
vif['features']=x_data2_2.columns
vif
```

```python
sns.pairplot(x_data2_2)
plt.show()
```

```python
x_data2_1_ = sm.add_constant(x_data2_1,has_constant='add')
x_data2_2_ = sm.add_constant(x_data2_2,has_constant='add')
```

```python
model_vif1 = sm.OLS(target,x_data2_1_)
fitted_model_vif1 = model_vif1.fit()
model_vif2 = sm.OLS(target,x_data2_2_)
fitted_model_vif2 = model_vif2.fit()
```

```python
fitted_model_vif1.summary()
# NOX제거
# NOX제거 전과비교했을때 R^2가 크게 변함이 없었다. -> NOX가 영향을 끼치는건 적었다.
```

```python
fitted_model_vif2.summary()
# NOX, RM 제거
# NOX만 제거한것에 비해 R^2가 상당히 줄었다. 그리고 CRIM과 INDUS의 선형회귀계수가 증가한다
# RM은 영향을 끼친다 제거하면 안된다
# AIC ,BIC도 늘어났다. 얘네는 작을수록 좋은것
```

# 학습 / 검증데이터 분할

```python
from sklearn.model_selection import train_test_split
X = x_data2_
Y = target
train_x, test_x, train_y, test_y = train_test_split(X,Y,train_size=0.7, test_size=0.3,random_state=1)
print(train_x.shape, test_x.shape, train_y.shape, test_y.shape)
```

```python
#  train_x 회귀모델 적합
train_x.head()
fit_1 = sm.OLS(train_y, train_x)
fit_1 = fit_1.fit()
```

```python
# 검증데이터에 대한 예측값과 true 값 비교
plt.plot(np.array(fit_1.predict(test_x)),label='pred')
plt.plot(np.array(test_y),label='true')
plt.legend()
plt.show()

# 자세한 값은 튀긴하지만 어느정도 정황은 맞춘다
```

이상변수를 제거하고 학습 검증데이터

```python
# 이상변수를 제거한 것에 대한 학습 검증데이터 분할
# NOX제거하고 상수한 추가한 데이터
X = x_data2_1_
Y = target
train_x2, test_x2, train_y2, test_y2 = train_test_split(X,Y,train_size=0.7, test_size=0.3,random_state=1)

# NOX, RM제거하고 상수한 추가한 데이터
X = x_data2_2_
Y = target
train_x3, test_x3, train_y3, test_y3 = train_test_split(X,Y,train_size=0.7, test_size=0.3,random_state=1)
```

```python
fit_2 = sm.OLS(train_y2, train_x2)
fit_2 = fit_2.fit()
fit_3 = sm.OLS(train_y3, train_x3)
fit_3 = fit_3.fit()
```

```python
# 검증데이터에 대한 예측값과 true 값 비교
plt.plot(np.array(fit_2.predict(test_x2)),label='pred2')
plt.plot(np.array(fit_3.predict(test_x3)),label='pred3')
plt.plot(np.array(test_y2),label='true')
plt.legend()
plt.show()

# 자세한 값은 튀긴하지만 어느정도 정황은 맞춘다
# 변수하나만 제거한게 조금 낫다
```

```python
# 그래프가 겹쳐서 잘 안보이므로 차이를 그려보자
plt.plot(np.array(test_y2['Target'] - fit_1.predict(test_x)),label='pred_full')
plt.plot(np.array(test_y2['Target'] - fit_2.predict(test_x2)),label='pred_vif1')
plt.plot(np.array(test_y2['Target'] - fit_3.predict(test_x3)),label='pred_vif2')
plt.legend()
plt.show()

# 주황색( 변수1개제거 )가 잔차가 제일 적다
```
