# 단순선형회귀분석

```python
import os
import pandas as pd
import numpy as np
import statsmodels.api as sm
```

```python
# 현재경로 확인
os.getcwd()
```

```python
# 데이터 불러오기
boston = pd.read_csv('./Boston_house.csv')
```

```python
boston.head()
```

```python
# target값 제외한데이터만 뽑기
boston_data = boston.drop(['Target'],axis=1)
```

```python
# data 통계 뽑기
boston_data.describe()
```

## CRIM/RM/B 세개의 변수로 각각 단순 선형 회귀분석

```python
# 변수 설정 target / CRIM/ RM / B
target = boston['Target']
crime = boston['CRIM']
rm = boston['RM']
black = boston['B']
```

```python
target
```

## target - crime 선형회귀분석

```python
# crime 변수에 상수항 추가하기
crime1 = sm.add_constant(crime,has_constant="add")
crime1
```

```python
# sm.OLS 적합시키기
model1 = sm.OLS(target,crime1)
fitted_model1 = model1.fit()
```

```python
fitted_model1.summary()
```

```python
# 회귀 계수 출력
fitted_model1.params
```

## y_hat = beta0 + beta1*X 계산해보기

```python
# 회귀 계수 * 데이터(X)
np.dot(crime1,fitted_model1.params)
```

```python
# predict함수를 통해 yhat구하기
pred1 = fitted_model1.predict(crime1)
```

```python
# 직접구한 yhat과 predict함수를 통해 구현한 yhat차이
# 차가 0임을 알수있다
np.dot(crime1,fitted_model1.params) - pred1
```

```python
import matplotlib.pyplot as plt
plt.scatter(crime,target,label='data')  # 원래 값
plt.plot(crime,pred1,label='result')  # result그림처럼 나와야 잘 해석된 값
plt.legend()
plt.show()
```

```python
# residual 시각화
fitted_model1.resid.plot()
plt.xlabel('residual_number')
plt.show()
#잔차가 일정하지않다 범죄율 예측하기엔 부족하다
```

```python
# 잔차의 합 계산하기
np.sum(fitted_model1.resid)
# 거의 0에 가까움 0이라고 볼수있음 . sse가 최소가 되도록 선을 정하는 것이기 때문에 0임이 당연하다
```

## 나머지 변수 ( rm, b)에 대해 단순회귀분석 적용시키기

```python
fitted_model2.summary()
## r-sqaured value가 클수록 집값이 큰 영향을 미친다
```

```python
fitted_model3.summary()
```

```python
# 각각 yhat 예측하기
pred2 = fitted_model2.predict(rm1)
pred3 = fitted_model3.predict(black1)
```

```python
import matplotlib.pyplot as plt
plt.scatter(rm,target,label='data')  # 원래 값
plt.plot(rm,pred2,label='result')  # result그림처럼 나와야 잘 해석된 값
plt.legend()
plt.show()
```

```python
fitted_model2.resid.plot()
```

```python
import matplotlib.pyplot as plt
plt.scatter(black,target,label='data')  # 원래 값
plt.plot(black,pred3,label='result')  # result그림처럼 나와야 잘 해석된 값
plt.legend()
plt.show()
```

```python
# 세 모델의 잔차 비교
fitted_model1.resid.plot(label='crime')
fitted_model2.resid.plot(label='rm')
fitted_model3.resid.plot(label='black')
plt.legend()
# 주황색 rm이 0에 가깝게 분포함을 볼수있음 
# 따라서 집값 예측에 간단하게 rm값을 사용할 수 있음 
```

# 다중선형회귀분석

```python
boston_data = boston.drop(['Target'],axis=1)
boston_data.head()
```

```python
target = boston[['Target']]
crim = boston[['CRIM']]
rm = boston[['RM']]
lstat = boston[['LSTAT']]
```

```python
crim = sm.add_constant(crim, has_constant='add')
rm = sm.add_constant(rm, has_constant='add')
lstat = sm.add_constant(lstat, has_constant='add')
```

```python
model1 = sm.OLS(target, crim)
model2 = sm.OLS(target, rm)
model3 = sm.OLS(target, lstat)

fitted_model1 = model1.fit()
fitted_model2 = model2.fit()
fitted_model3 = model3.fit()
```

```python
fitted_model1.summary()
```

```python
fitted_model2.summary()
```

```python
fitted_model3.summary()
```

```python
#  r squre를 보고 lstat변수가 가장 영향이 많다라고 판단했음
```

# 다중선형회귀분석
### crim,rm,lstat 세개의 변수를 통해 다중회귀적합

```python
# boston data에서 crim,rm,lstat 변수만 뽑아오기
x_data = boston[['CRIM','RM','LSTAT']]
x_data.head()
```

```python
# 상수항 추가
x_data1 = sm.add_constant(x_data,has_constant='add')
```

```python
# 회귀모델 적합
multi_model = sm.OLS(target,x_data1)
fitted_multi_model = multi_model.fit()
```

```python
# summary
fitted_multi_model.summary()
```

## 단순선형회귀모델의 회귀계수와 비교

```python
# 단순선형회귀분석모델의 회귀계수
print(fitted_model1.params)
print(fitted_model2.params)
print(fitted_model3.params)
```

```python
# 다중선형회귀분석모델의 회귀계수
print(fitted_multi_model.params)
# 단순일때 crim은 -0.41이었는데 다중일때 -0.1로 변함 
# 나머지 변수들도 변함 
```

### R-sqaure이 뜻하는것은 y의 총변동에 대해서 x가 설명하는 변동의 비율
선형회귀분석에서는 x가 y의 변동성을 얼마나 잘 설명하냐가 중요했는데
변수가 여러개일 때 설명하는 변동성을 중복으로 가져갈 수 없다.
3개의 변수를 넣엇을때 R^2이 0.646이 되었다.
3번째 변수가 0.5정도였다.
그 이야기는 3개의 변수의 영향을 다 합쳤을때 줄어들기때문에 3개의 변수가 어느정도 중복된다라는 것을 의미한다  -> 다중공선성이라고 한다.
다중공선성이 있지만 각각의 p-value가 매우 낮기 때문에 3개의 변수를 다 쓸수는 있다고 생각할 수 있다.

## 행렬연산을 통해 beta 구하기

```python
from numpy import linalg  # 행렬연산을 통해 beta구하기
# 다중회귀에서 beta를 구할 때는 행렬로 계산했었음
# beta = (X'X)-1X'y

ba = linalg.inv(np.dot(x_data1.T,x_data1))
# ba : (X'X)-1

print(fitted_multi_model.params)
np.dot(np.dot(ba,x_data1.T),target)
# 직접구한값과 비교해보면 일치한다
```

```python
# y_hat구하기
pred4 = fitted_multi_model.predict(x_data1)
```

```python
import matplotlib.pyplot as plt
fitted_multi_model.resid.plot()
plt.show()
```

```python
fitted_model1.resid.plot(label='crim')
fitted_model2.resid.plot(label='rm')
fitted_model3.resid.plot(label='lstat')
fitted_multi_model.resid.plot(label='full')
plt.legend()
plt.show()

# 빨간색 full변수의 잔차가 가장 적음을 알 수 있다.
# 변수가 많아지면 많아질수록  R-sqare은 증가하고 학습데이터의 잔차는 줄어든다
# 마냥 변수가 많아진다고 좋은게 아니다. 변수들을 검증해야한다.
```
