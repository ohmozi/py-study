# k-Nearest Neighborgood algorithm

# 1. 데이터, 모듈 불러오기 및 kNN피팅 방법

- 함수 불러오기

```python
from sklearn import neighbors, datasets
```

```python
import numpy as np
import matplotlib.pyplot as plt
from matplotlib.colors import ListedColormap
```

```python
iris = datasets.load_iris()

X = iris.data[:,:2]
y = iris.target
```

```python
clf = neighbors.KNeighborsClassifier(5)  # k의 개수 넣어주기
clf.fit(X, y)
```

```python
y_pred = clf.predict(X)
```

```python
from sklearn.metrics import confusion_matrix
```

```python
confusion_matrix(y, y_pred)
```

# 2. Cross-validation을 활용한 최적의 k 찾기

```python
from sklearn.model_selection import cross_val_score
# skelearn.cross_validation 이 최신버전에서 model_selection으로 바뀜
```

- CV진행

```python
k_range = range(1,100)
k_scores=[]

for k in k_range:
    knn = neighbors.KNeighborsClassifier(k)   # ㅏ값을 변경해가며 KNN진행
    scores=cross_val_score(knn, X, y, cv=10, scoring="accuracy")  
    #print(scores)
    # 임의의 k에 대하여 10fold로 진행
    # 10fold마다 cross_val_score가 10번 나옴
    k_scores.append(scores.mean())
    # scores에 10번의 평균을 k_score에 저장
```

```python
plt.plot(k_range, k_scores)
plt.xlabel('k_range')
plt.ylabel('k_scores')
plt.show()
```

```python
# 정확도가 제일 높은 k
np.argmax(k_scores)
```

# 2. Weight를 준 kNN

```python
np.random.seed(0)
X = np.sort(5 * np.random.rand(40, 1), axis=0)
# 40*1행으로 0~1사이 실수 생성,  axis=0으로 정렬
T = np.linspace(0, 5, 500)[:, np.newaxis]
# 테스트 데이터
y = np.sin(X).ravel()
# y를 sin파에 대입
y[::5] += 1 * (0.5 - np.random.rand(8))
# 중간중간 이상치 생성
```

```python
knn = neighbors.KNeighborsRegressor(5)
y_ = knn.fit(X, y).predict(T)
#y_pred2 = y_.predict(T)

# 범주형 변수일때는 KNeighborsClassifier(k)
# 연속형 변수일때는 KNeighborsRegressor(k)
```

```python
n_neighbors = 5

for i, weights in enumerate(['uniform', 'distance']):
    knn = neighbors.KNeighborsRegressor(n_neighbors, weights = weights)
    y_ = knn.fit(X, y).predict(T)
    
    plt.subplot(2,1, i+1)
    plt.scatter(X, y, c='k', label='data')
    plt.plot(T, y_, c='g', label='prediction')
    plt.axis('tight')
    plt.legend()
    plt.title("KNeighborsRegressor (k = %i, wieghts = '%s')" % (n_neighbors, weights))
    
plt.tight_layout()
plt.show()
```

- 항상 distance weight를 주는게 좋은 모델은 아니지만 상황에 따라 확인하기
