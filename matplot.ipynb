```python
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

## 그래프 데이터 생성

```python
x = np.linspace(0,10,11)
y = x**2 + x + 2 + np.random.randn(11) # 노이즈를 준것

print(x)
print(y)
```

```python
plt.plot(x,y)
```

```python
plt.scatter(x,y)
```

## 그래프에 주석추가

```python
plt.xlabel('x value')
plt.ylabel('y value')
plt.title('x-y title')
plt.grid(True)

plt.plot(x,y)
```

## x,y축 범위 지정

```python
plt.xlabel('x value')
plt.ylabel('y value')
plt.title('x-y title')
plt.grid(True)

plt.xlim(0,20)
plt.ylim(0,200)

plt.plot(x,y)
```

```python
plt.plot(x,y,'#fddf3f')
```

```python
plt.plot(x,y,'-.')
```

```python
plt.plot(x,y,'g^')
```

```python
plt.plot(x,y,'r:')
```

```python
plt.plot(x,y,'-.',linewidth=5)
```

## subplot

```python
plt.subplot(2,2,1)
plt.plot(x,y,'r')
plt.subplot(2,2,2)
plt.plot(x,y,'b')
plt.subplot(2,2,3)
plt.plot(x,y,'r^')
plt.subplot(2,2,4)
plt.plot(x,y,'b^')
```

```python
x = np.random.randint(0,101,100)
print(x)
```

```python
total = int(1e7)
points = np.random.rand(total,2)
# 천만개의 점 생성하기
points

in_points = points[np.sum(points**2, axis=1)<1]
# np.sum(np.sum(points**2,axis=1)<1)
print(len(in_points))

print(4*len(in_points)/total)
```
