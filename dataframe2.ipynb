```python
import pandas as pd
import numpy as np
```

```python
df = pd.read_csv('./train.csv')
df.head()
```

## transform 함수
- groupby 후 transfrom 함수를 사용하면 원래의 index를 유지한 상태로 통계함수를 적용
- 전체 데이터의 집계가 아닌 각 그룹에서의 집계를 계산
- 따라서 새로 생성된 데이터를 원본 dataframe과 합치기 쉬움

```python
df.groupby('Pclass').mean()
```

```python
df.groupby('Pclass').transform(np.mean)
# 원본에 각 그룹별 값을 넣는다  
# 위의 3개의 평균값이 아래 각각 중복해서 들어가있는 것을 확인할 수 있다
```

```python
df.groupby('Pclass').transform(np.mean)['Age']
```

```python
df['Age2'] = df.groupby('Pclass').transform(np.mean)['Age']
df.head()
```

```python
df.groupby(['Pclass','Sex']).mean()
```

```python
df.groupby(['Pclass','Sex']).transform(np.mean)['Survived']
```

```python
df['Survived2'] = df.groupby(['Pclass','Sex']).transform(np.mean)['Survived']
df
```

# pivot

```python
df = pd.DataFrame({
    '지역' : ['서울','서울','서울','경기','경기','부산','서울','서울','부산','경기','경기','경기'],
    '요일' : ['월','화','수','월','화','월','목','금','화','수','목','금'],
    '강수량' : [100, 80, 1000, 200, 200, 100, 50, 100, 200, 100, 50, 100],
    '강수확률' : [80, 70, 90, 10, 20, 30, 50, 90, 20, 80, 50, 10]
    })
df
```

## pivot
- dataframe의 형태를 변경
- 인덱스, 컬럼, 데이터로 명시
- 중복되는 인덱스,컬럼이 존재하면 에러남

```python
# index로 지정할 컬럼, 컬럼으로 지정할 컬럼을 명시해주면된다
# pivot(index,column,value)
df.pivot('지역','요일')
```

```python
df.pivot('요일','지역')
```

## pivot_table
- 기능적으로 pivot과 동일
- pivot과 차이점
    - 중복되는 모호한 값이 있을 경우, aggregation 함수 사용하여 값을 채움

```python
# 서울, 월요일 값이 중복될때  오류를 발생시키지 않고 aggfunc으로 에러 처리 ( 지금은 평균값 삽입 )
pd.pivot_table(df,index='요일',columns='지역',aggfunc=np.mean)
```

## stack & unstack
- stack : 컬럼 레벨에서 인덱스레벨로 dataframe 변경
    - 즉, 데이터를 쌓아올리는 개념으로 이해하면 쉬움
- unstack : 인덱스 레벨에서 컬럼레벨로

```python
df
```

```python
new_df = df.set_index(['지역','요일'])
new_df
```

```python
new_df.unstack(0)
# 지역인덱스를 컬럼으로 
```

```python
new_df.unstack(0).stack(0)
```

## concat

```python
df1 = pd.DataFrame({'key1' : np.arange(10), 'value1' : np.random.randn(10)})
```

```python
df2 = pd.DataFrame({'key1' : np.arange(10), 'value1' : np.random.randn(10)})
```

```python
df1
```

```python
df2
```

```python
pd.concat([df1,df2])
# 인덱스가 유지됨
# 인덱스를 무시하고 싶으면 ignore_index=True
```

```python
pd.concat([df1,df2],ignore_index=True)
```

```python
pd.concat([df1,df2],axis=1)
```

### 컬럼명이 다를경우

```python
df3 = pd.DataFrame({'key3' : np.arange(10), 'value3' : np.random.randn(10)})
```
