## NaN 처리 방법
- 데이터에서 삭제
    - dropna 함수
- 다른 값으로 치환
    - fillna 함수

```python
print(len(train_data))
print(len(train_data.dropna()))
```

```python
train_data.dropna(subset=['Age'])
# 나이값에 NAn이 있으면 드랍
```

```python
train_data.dropna(axis=1)
# 값중에 하나라도 nan이 있으면 컬럼 자체를 삭제
```

```python
len(train_data.dropna(axis=0))
# 값중에 하나라도 nan이 있으면 로우 자체를 삭제
```

- NaN 값 대체하기
    - 평균으로 대체하기
    - 생존자/ 사망자 별 평균으로 대처하기

```python
train_data['Age'].fillna(train_data['Age'].mean())
```

```python
# 생존자 나이 평균
meanl = train_data[train_data['Survived']==1]['Age'].mean()

# 사망자 나이 평균
meand = train_data[train_data['Survived']==0]['Age'].mean()

print(meanl)
print(meand)
```

```python
train_data[train_data['Survived']==1]['Age'].fillna(meanl) 
```

```python
train_data[train_data['Survived']==0]['Age'].fillna(meand)
```

```python
train_data.loc[train_data['Survived']==1,'Age']
```

```python
train_data.loc[train_data['Survived']==1,'Age'] = train_data[train_data['Survived']==1]['Age'].fillna(meanl)
# 살아있는 사람의 나이를 가져오고  / 살아있는 사람중 나이 컬럼에 NaN값을 나이 평균으로 대체
train_data.loc[train_data['Survived']==0,'Age'] = train_data[train_data['Survived']==0]['Age'].fillna(meand)
```
